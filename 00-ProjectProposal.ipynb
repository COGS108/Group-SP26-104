## Research Question

Do individuals who use wearable fitness trackers report greater alignment between perceived and actual physical activity levels, and does this vary by socioeconomic status and health self efficacy? 

## Background and Prior Work

The proliferation of consumer-grade wearable fitness trackers (devices like the Fitbit, Apple Watch, and Garmin) has introduced a new paradigm in personal health monitoring. Marketed on the promise of allowing users to better understand their levels of physical activity with scrutiny previously inaccessible, they are now closing the gap between how active a person thinks they are compared to the reality of their activity. As such, these devices now occupy a central role in the conversation of public health. However, it is questionable that the relationship between tracker use, self-perception, and behavioral changes are as simple as is provided by the marketing labels placed on such tools, and the benefits communicated are not distributed evenly across socioeconomic groups.

A foundational concern in this area is whether wearable devices reliably measure physical activity in the first place. Research evaluating major commercial trackers found that while metrics like heart rate, step count, and sleep duration were measured with relatively high accuracy, energy expenditure (calories burned) showed significantly higher error rates, with mean absolute percentage error reaching as high as 0.44 across devices. This matters because caloric output is one of the primary metrics users rely on when assessing whether their activity levels are "adequate." A closer analysis of the issue also reinforces the worry. For devices like the Fitbit Flex, the error in measuring time spent in moderate-to-vigorous physical activity ranged dramatically: from 7% to 74% depending on the study. This indicats substantial variability in actual accuracy. These inconsistencies suggest that the alignment between perceived and actual activity a tracker produces may be artifact-dependent, not just behavior-dependent, and this is a genuine threat to the validity that any study in this space must account for.<a name="cite_ref-1"></a><sup>1</sup>

The theoretical backbone of this research draws on Bandura's Social Cognitive Theory (SCT), particularly the construct of self-efficacy. Self-efficacy is defined as an individual's belief in their capacity to perform a behavior successfully. SCT proposes that self-efficacy is the most influential determinant of physical activity behavior, with high self-efficacy leading to stronger expectations, more concrete goals, and greater persistence when faced with obstacles. Wearable trackers are, in theory, powerful self-efficacy tools: they provide real-time feedback, create visible progress toward goals, and reinforce mastery in activities. According to current data, trackers have been shown to increase participant self-efficacy by providing real, visible reminders of commitment to activity and immediate feedback on a person's progression, though it might be noted that some of this evidence comes from intervention studies with supported environments (wellness groups, physician-led talks, etc), which might be seen to put into question whether these effects generalize to everyday use. Nonetheless, it is a reflection of real relationships between opinion, optimism, and generalizable evidence.<a name="cite_ref-2"></a><sup>2</sup>

One of the most interesting but underexplored findings in this literature is that the perception of physical activity may carry independent health significance beyond actual behavior. Epidemiological research drawing on three nationally representative samples found that individuals who perceived themselves as less active than their peers had a mortality risk up to 72% higher over a 21-year follow-up period, even after controlling for actual physical activity levels. This suggests that how accurately one perceives their own activity. It isn't just the activity someone performs, but also their personal thoughts on their well-being that hold real health implications, and that tools capable of correcting or distorting that perception (like the wearables studied in this proposal) deserve serious empirical scrutiny.<a name="cite_ref-3"></a><sup>3</sup>
An often overlooked dimension of wearable technology is equity. That fitness trackers democratize health data is an assumption that is not often considered or supported. When participant demographics in a large national health research program (NIH's 'All of Us') were examined, it was found that individuals who contributed fitbit data showed a notable reduction in racial and socioeconomic diversity as compared to the broader pool of overall participants, with cost, awareness, and language found to be the key barriers when it came to contribution. Additionally, sociological research shows that physical activity trackers are used more frequently from individuals from more affluent backgrounds, which reflect a potential concern that technological health tools may reinforce rather than reduce existing health inequalities. This idea, mostly absent from device-focussed accuracy studies, shows an important alternative view that must be considered in any model that claims to understand the efficacy of health trackers at a national level.<a name="cite_ref-4"></a><sup>4</sup>
The current literature establishes 3 main tensions. For one, trackers are assumed to improve the alignment between perceived activity and actual activity, yet device accuracy varies significantly by metric and context. Second, self-efficacy is a hard to define mediator for behavioral change, though tracker based self-efficacy gains might be context-dependent, and finally, socioeconomic status shapes both tracker access and the conditions in which feedback loops are effective. No study using survey data has simultaneously tied together all three of these factors in a general adult population. This is the gap this proposal will address.

## Hypothesis


We predict that individuals who use wearable fitness trackers will report greater alignment between their perceived and actual physical activity levels compared to non-users. However, we expect this benefit to be significantly moderated by socioeconomic status (SES) and health self-efficacy. Specifically, we anticipate that higher-SES individuals with greater health self-efficacy will demonstrate the strongest alignment between self perceived and actual activity, while lower-SES individuals will show weaker alignment even when tracker use is present. This reflects disparities in device access, engagement quality, and the social conditions under which feedback loops are effective. This prediction is mainly founded on background studies that show how tracker benefits are most pronounced in supported, resource-rich environments, and that cost and awareness barriers disproportionately limit meaningful tracker engagement.

## Data

<h3>Ideal Dataset</h3>

The ideal dataset for this project would be a large scale cross sectional survey of consenting U.S. adults capturing variables such as:

- Wearable tracker use - binary variable (yes/no), frequency (daily, weekly, rarely), monitored metrics (steps, calories, heartrate)
- Perceived physical activity - self-rated activity level
- Actual physical activity - validated self-report measure or accelerometer data
- Socioeconomic status (SES) - household income bracket, educational attainment, employment status
- Demographic / controls - age, gender, race/ethnicity, geographic region, BMI

**Observations needed:** 1,000-1,500 respondents or more would be needed to detect the hypothesized effects with sufficient statistical reliability.

**Collection method:** Data would be collected by a random sample survey. The survey should be nationally representative and is adminstered online and by phone. Recruitment would target adult age groups, diverse income, and a wide range of SES factors.

**Storage/organization:** Data would be stored in a CSV format, with one row per respondent and columns for each variable. Personally identifiable information would be removed prior to analysis. 

<h3>Potential Datasets</h3>

1. Health Information National Trends Survey (HINTS)
- Available at: https://hints.cancer.gov/data/default.aspx
- publicly available, no application required
- HINTS is a nationally representative survey administered by the National Cancer Institute that specifically tracks a plethora of health categories. Importantly, in the more recent cycles (since HINTS 5 cycle 3), this includes use of health technology, including wearable devices and fitness apps. It contains variables on fitness tracker ownership and use frequency, perceived health behaviors, health self-efficacy, and sociodemographic characteristics. This makes HINTS the closest real-world match to the ideal dataset for this project. Data is available for immediate download in CSV and SAS formats as well. This would likely serve as the primary dataset for the project.
2. Behavioral Risk Factor Surveillance System (BRFSS)
- Available at: https://www.cdc.gov/brfss/data_documentation/index.htm
- publicly available, no application required
- The BRFSS is an annual CDC telephone survey of over 400,000 U.S. adults, making it one of the largest health-related survey datasets in the world. It includes self-reported physical activity levels, demographic variables, income, education, and health status measures highly relevant to this project. 
It does not directly ask about wearable tracker ownership, physical activity perception and actual activity variables can be operationalized from existing questions. Its enormous sample size makes it exceptionally well-suited for subgroup and moderation analysis by SES. Data is freely downloadable in SAS XPT format and can be converted to CSV for use in Python or R.
3. National Health Interview Survey (NHIS)
- Available at: https://www.cdc.gov/nchs/nhis/documentation/?CDC_AAref_Val=https://www.cdc.gov/nchs/nhis/data--questionnaires-documentation.htm 
- publicly available, no application required
- The NHIS is a large household-based survey conducted annually by the CDC's National Center for Health Statistics, covering over 25,000 households per year. It includes detailed physical activity frequency and duration variables, self-rated health measures, income, education, and insurance status as SES indicators. This set also does not include a direct wearable variable, but it has the richest set of physical activity measurement of the three sets here, and will be used to validate or supplement foundational findings in HINTS.
