## Research Question

Do individuals who use wearable fitness trackers report greater alignment between perceived and actual physical activity levels, and does this vary by socioeconomic status and health self efficacy? Socioeconomic status will be measured by household income and educational attainment) and health self-efficacy will be measured by self rated confidence in maintaining an active lifestyle. This is primarily a statistical inference task, using logistic and multiple linear regression to estimate the association between tracker use and perception-reality alignment, with SES and self-efficacy included as moderating variables. Secondary subgroup analyses will assess whether the effect of tracker use on alignment differs significantly across income brackets.

The ideal dataset for this project would be a large scale cross sectional survey of U.S. adults capturing variables such as:
- Wearable tracker use - binary variable (yes/no), frequency (daily, weekly, rarely), metrics that user monitors (steps, calories, heartrate)
- Perceived physical activity - self-rated activity level
- Actual physical activity - validated self-report measure or accelerometer data
- Socioeconomic status (SES) - household income bracket, educational attainment, employment status
- Demographic / controls - age, gender, race/ethnicity, geographic region, BMI

**Observations needed:** 1,000 - 1,500 respondents or more would be needed to detect effect sizes with sufficient statistical power. This is important particularily for subgroup analyses by socioeconomic status. 

**Collection method:** Data would be collected by a random sample survey. The survey should be nationally representative and is adminstered online and by phone. Recruitmnet would target all age groups, diverse income, with additional focus on lower SES respondents to ensure adequate representation. 

**Storage/organization:** Data would be stored in a CSV format, with one row per respondent and columns for each variable. Personally identifiable information would be removed prior to analysis.
